# Сделать CSV где 1 материал только

```python
import pandas as pd
import re

df = pd.read_csv(r""+path+"\\"+filee+"_"+mater+".csv")
column = df['Изображения'].tolist()
for i in range(len(column)):
    column[i] = column[i].translate(str.maketrans({"'": ''}))
    column[i] = re.sub('url: ', '', column[i])
    column[i] = column[i].split(',')

h = []

for i in range(len(column)):
    h.append(column[i][0])

    
for ii in range(len(h)):
    h[ii] = h[ii][59:]+str('.jpg')

df = pd.read_csv(r""+path+"\\"+filee+"_"+mater+".csv")
column2 = len(column) * [mater]


data = dict(file = h, label=column2)
df = pd.DataFrame(data)
df.to_csv(r""+path+"\\"+mater+'.csv', encoding="utf-8", index=False)
```

# Сделать TXT

```python
import pandas as pd

df = pd.read_csv(r""+path+"\\"+filee+"_"+mater+".csv")
column = df['Изображения'].tolist()

gg = []
for i in range(len(column)):
    clm = column[i].split(',')
    try:
        gg.append(clm[1])
    except:
        print(1)

with open(r""+path+"\\"+mater+'.txt', 'w', encoding="utf-8") as filehandle: 
    for listitem in gg: 
        filehandle.write('%s\n' % listitem)
```
